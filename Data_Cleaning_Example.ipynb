```python
 import pandas as pd
 data = {
    'Campus': ['NorthTech', 'GreenVille ', 'GreenVille ', None, 'EcoPark', 'NorthTech',
 'GreenVille ', 'EcoPark'],
    'Waste_Recycled_kg': [1200, 1500, None, 1300, 1600, 1200, 1500, 1600],
    'Electric_Vehicles': [5, 7, 6, 8, None, 5, 7, 5],
    'Sustainability_Score': [80, 85, 78, 90, 95, None, 85, 95]
 }
 df = pd.DataFrame(data)
 df.to_csv("sustainability_survey.csv", index=False)
 df
```

```python
df['Campus'].fillna('Unknow',inplace=True)
df['Waste_Recycled_kg'].fillna(df['Waste_Recycled_kg'].mean(),inplace=True)
df['Electric_Vehicles'].fillna(df['Electric_Vehicles'].mean(),inplace=True)
df['Sustainability_Score'].fillna(df['Sustainability_Score'].mode()[0],inplace=True)
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['Campus']=df['Campus'].str.lower().str.strip()
df['Campus']
```

```python
from scipy.stats import zscore
df['z_score_energy']=zscore(df['Waste_Recycled_kg'])
df=df[(df['z_score_energy']>-3)&(df['z_score_energy']<3)]
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,StandardScaler
label_encoder=LabelEncoder()
df['Campus_Label']=label_encoder.fit_transform(df['Campus'])
```

```python
df
```

```python
from scipy.stats import zscore
df['z_score_energy']=zscore(df['Sustainability_Score'])
df=df[(df['z_score_energy']>-3)&(df['z_score_energy']<3)]
```

```python
df
```

```python
label_encoder=LabelEncoder()
df['Campus_Label']=label_encoder.fit_transform(df['Campus'])
```

```python
df
```

```python
df_campus_encoded=pd.get_dummies(df['Campus'],prefix='Campus',dtype=int)
df=pd.concat([df,df_campus_encoded],axis=1)
df
```

```python
scaler=MinMaxScaler()
df[['Waste_Recycled_kg_Norm','Electric_Vehicles_Norm','Sustainability_Score_Norm']]=scaler.fit_transform(df[['Waste_Recycled_kg','Electric_Vehicles','Sustainability_Score']])
```

```python
df
```

```python
standard_scaler=StandardScaler()
scaled_data = standard_scaler.fit_transform(df[['Waste_Recycled_kg','Electric_Vehicles','Sustainability_Score']])
df['Waste_Recycled_kg_Std'] = scaled_data[:, 0]
df['Electric_Vehicles_Std'] = scaled_data[:, 1]
df['Sustainability_Score_Std'] = scaled_data[:, 2]
df
```
