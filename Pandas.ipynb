### .data manuplation in pandas

#### step1 create a dataset

```python
import pandas as pd
data ={
    'city':['banglore','chennai','hydarbad','dbpur','delhi'],
    'solar_units':[120,85,60,150,90],
    'wind_units':[20,50,80,40,85],
    'solar_consumer':[1000,4000,3000,54000,3002]
}
df = pd.DataFrame(data)
df
```

#### basic data exploration

```python
print(df.shape)
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.columns)
print(df.dtypes)
```

```python
print(df.head())
print(df.tail())
```

```python
print(df.head(2))
print(df.tail(2))
```

#### step3 data selection and filtering

```python
print(df['city'])
```

```python
print(df[['city','solar_units']])
```

```python
print(df[df["solar_units"] > 100])
```

```python
print(df[df["solar_consumer"] >= 100 & 6000])
```

#### step4 data manuplation

```python
df['total_units']= df['solar_units']+df['wind_units']
df
```

```python
df['units_per_1000_consumer']= df['total_units'] / df['total_units']
df
```

```python
df.rename(columns={'city': 'city_name'},inplace=True)
df
```

#### step5 sorting aggregation

```python
print(df.sort_values(by='solar_units',ascending=False))
```

```python
print(df.sort_values(by='wind_units',ascending=True))
```

```python
grouped = df.groupby('solar_units').sum()
df
```

```python
grouped = df.groupby('wind_units').sum()
df
```

```python
wind_units = df['wind_units'].sum()
df
```

#### step6 handling missing values

```python
df.loc[2,'solar_units']=None
df
```

```python
print(df.isnull())
```

```python
df.isnull().sum()
```

```python
df['solar_units'].fillna(df['solar_units'])
```

```python
df['solar_units'].fillna(df['solar_units'].mean(),inplace=True)
```

#### step7 merging & joininng

```python
energy_type =  {
    'city_name':['banglore','chennai','hydarbad','dbpur','delhi'],
    'primary_source':['solar','wind','solar','wind','solar']
}
energy_type_df=pd.DataFrame(energy_type)
df
```

```python
merged_df = pd.merge(df,energy_type_df,on='city_name')
print(merged_df)
```
