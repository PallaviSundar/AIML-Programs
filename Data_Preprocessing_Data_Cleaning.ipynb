### Data clening and green ai dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,StandardScaler
from scipy.stats import zscore

# Step 1: Create Dataset
data = {
    'College': ['GreenTech ', 'EcoWorld', 'GreenTech ', 'BioCampus', 'EcoWorld', None, 'EcoWorld', 'GreenTech '],
    'Solar_Panels': [150, 200, np.nan, 100, 250, 180, 200, 150],
    'Energy_Saved_kWh': [3000, 4500, 4000, None, 5000, 4200, 4500, 3000],
    'Green_Projects': [5, 7, 6, 4, np.nan, 5, 7, 5],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Bangalore', 'Mumbai', 'Delhi']
}
df = pd.DataFrame(data)
df
```

### step2 missing value handling

```python
df['College'].fillna('Unknow',inplace=True)
df['Solar_Panels'].fillna(df['Solar_Panels'].mean(),inplace=True)
df['Energy_Saved_kWh'].fillna(df['Energy_Saved_kWh'].mean(),inplace=True)
df['Green_Projects'].fillna(df['Green_Projects'].mode()[0],inplace=True)
df
```

#### step3

```python
df_cleaned=df.dropna()
df_cleaned
```

#### step4 remove duplicate

```python
df = df.drop_duplicates()
df
```

#### step5 Standardize text

```python
df['College']=df['College'].str.lower().str.strip()
df['College']
```

#### step6 strip whitespace

```python
df['City']=df['City'].str.strip()
df['City']
```

#### step7 encoding

```python
label_encoder=LabelEncoder()
df['College_Label']=label_encoder.fit_transform(df['College'])
```

```python
df
```

#### step7 one hot encoding

```python
import pandas as pd
df_city_encoded=pd.get_dummies(df['City'],prefix='City',dtype=int)
df=pd.concat([df,df_city_encoded],axis=1)
```

```python
df
```

#### step8 Scalling

```python
scaler=MinMaxScaler()
df[['Solar_Panels_Norm','Energy_Saved_Norm']]=scaler.fit_transform(df[['Solar_Panels','Energy_Saved_kWh']])
```

```python
df
```

#### step9 Staderdization

```python
standard_scaler=StandardScaler()
df['Green_Projects_Std']=standard_scaler.fit_transform(df[['Green_Projects']])
df
```

#### step10 remove outliers on zscore

```python
df['z_score_energy']=zscore(df['Energy_Saved_kWh'])
df=df[(df['z_score_energy']>-3)&(df['z_score_energy']<3)]
df
```

#### step11 rename columns

```python
df=df.rename(columns={
    'College':'College_Name',
    'Solar_Panels':'Solar_Panel_Count',
    'Energy_Saved_kWh':'Energy_Saved_kWh',
    'Green_Projects':'Green_Project_Count',
})
df
```
