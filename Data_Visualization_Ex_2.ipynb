```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Technology':['solar','wind','hydro','Geothermal','biothermal'],
    'production_TWh':[450,600,850,90,120],
    'emission_reduced_MTCO2':[300,400,700,60,80],
    'cost_per_mwh':[35,30,50,45,40]
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['Technology'], df['production_TWh'], marker='o')
plt.title('line chart-Production by technology')
plt.xlabel('Technology')
plt.ylabel('Production (TWh)')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Technology'], df['production_TWh'])
plt.title('line chart-Production by technology')
plt.xlabel('Technology')
plt.ylabel('Production (TWh)')
plt.grid(True)
plt.show()
```

```python
plt.pie(df['cost_per_mwh'], labels=df['Technology'], autopct='%1.1f%%')
plt.title('pie chart - Cost distribution by Technology')
plt.show()
```

```python
plt.scatter(df['production_TWh'], df['cost_per_mwh'], color='blue')
plt.title('Scatter plot - Production vs cost')
plt.xlabel('production_TWh')
plt.ylabel('cost ($/MWh)')
plt.grid(True)
plt.show()
```

```python
df[['production_TWh', 'emission_reduced_MTCO2','cost_per_mwh']].plot(kind='box')
plt.title('box plot - Green AI metrics')
plt.show()
```
